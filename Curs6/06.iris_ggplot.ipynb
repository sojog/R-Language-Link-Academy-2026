```r
library(ggplot2)
```

```r
summary(iris)
```

# Plot gol

```r
ggplot(data=iris)
```

# Plot cu titlu si alte elemente

```r
ggplot(data=iris) + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala")
```

# AESTHETICS - coordonate

```r
ggplot(data=iris, aes(x=Sepal.Length, y= Sepal.Width)) + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala")
```

# Geometric objects

```r
ggplot(data=iris, aes(x=Sepal.Length, y= Sepal.Width)) + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_point()
```

# AESTHETICS - culori, marimi

```r
aes_obj = aes(x=Sepal.Length, y= Sepal.Width, color=Species)


ggplot(data=iris, aes_obj)  + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_point()
```

# Geometric objects - marime, transparenta

```r
aes_obj = aes(x=Sepal.Length, y= Sepal.Width, color=Species)


ggplot(data=iris, aes_obj)  + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_point(size=4)
```

```r
aes_obj = aes(x=Sepal.Length, y= Sepal.Width, color=Species)


ggplot(data=iris, aes_obj)  + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_point(size=4, alpha=0.5)
```

# Themes - teme

```r
ggplot(data=iris, aes(x=Sepal.Length, y= Sepal.Width, color=Species))  + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_point(size=4, alpha=0.5) + theme_minimal()
```

```r
ggplot(data=iris, aes(x=Sepal.Length, y= Sepal.Width, color=Species))  + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_point(size=4, alpha=0.5) + theme_dark()
```

# Linear regression

```r
library(dplyr)
```

```r
only_setosa <- iris %>% filter(Species=="setosa")
only_setosa %>% head(3)
```

```r
ggplot(data=only_setosa, aes(x=Sepal.Length, y= Sepal.Width, color=Species))  + labs(title = "Acesta este titlul plot-ului", x="Lungime sepala", y="Lățime sepala") + geom_smooth(method="lm")+ geom_point(size=4, alpha=0.5)
```
