```r


df <- read.csv("fish.csv")
```

# # 1. Creați un scatter plot cu Length3 pe axa X și Weight pe axa Y. Adăugați titlu și etichete pentru axe și theme_minimal().

# 

```r
ggplot(data=df, aes(x=Length3, y= Weight)) + labs(title = "Plotare pesti", x="Lungime", y="Greutate")
```

```r
ggplot(data=df, aes(x=Length3, y= Weight)) + labs(title = "Plotare pesti", x="Lungime", y="Greutate") + geom_point() + theme_minimal ()
```

#  2. Plecând de la Pct.1, colorează punctele după Species.
Setează alpha = 0.7 și size = 2 în geom_point().

```r

ggplot(data=df, aes(x=Length3, y= Weight, color=Species)) + labs(title = "Plotare pesti", x="Lungime", y="Greutate") + geom_point(size=2, alpha=0.7) + theme_minimal ()
```

# 3. Faceți un scatter Length3 vs Weight în care toate punctele sunt roșii (culoare fixă). În același grafic, păstrați alpha = 0.6.

```r
ggplot(data=df, aes(x=Length3, y= Weight)) + labs(title = "Plotare pesti", x="Lungime", y="Greutate") + geom_point(size=2, alpha=0.7) + theme_minimal ()
```

```r
COLOR = "red"

ggplot(data=df, aes(x=Length3, y= Weight)) + labs(title = "Plotare pesti", x="Lungime", y="Greutate") + geom_point(size=2, alpha=0.7, color=COLOR) + theme_minimal ()
```

# Linear regression - echivalent in Python modulul scipy (ulterior, scikit-learn)

```r
COLOR = "red"

ggplot(data=df, aes(x=Length3, y= Weight)) + labs(title = "Plotare pesti", x="Lungime", y="Greutate") + geom_smooth(method="lm") + geom_point(size=2, alpha=0.7, color=COLOR) + theme_minimal ()



```
