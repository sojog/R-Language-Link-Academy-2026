```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
BLUE = np.array([0, 43, 127])
YELLOW = np.array([252, 209, 22])
RED = np.array([206, 17, 38])
```

```python
SCALE = 1 * 100

WIDTH = 15 * SCALE
HEIGHT = 9 * SCALE
```

```python
steag_arr = np.ones((HEIGHT, WIDTH, 4), dtype=np.uint8)
```

```python
steag_arr[:,:,-1] = 255
steag_arr
```

```python
new_image = Image.fromarray(steag_arr)
new_image.save("steag_negru.png")
```
