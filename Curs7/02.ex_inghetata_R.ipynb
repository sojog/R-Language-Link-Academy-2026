# 2. Folosiți culoarea verde pentru vânzările din weekend și albastru în rest


### V1 - folosind plot clasic. 

```r
unique(df$zi_saptamana)
```

```r
weekend <- c('sambata','duminica')
weekend
```

```r
df_weekend <- df[df$zi_saptamana == 'sambata' |  df$zi_saptamana == 'duminica', ]
str(df_weekend)
```

```r
df_weekend <- df[df$zi_saptamana %in% weekend, ]
str(df_weekend)
```

```r
df_zile_lucratoare <- df[ ! (df$zi_saptamana %in% weekend), ]
str(df_zile_lucratoare)
```

```r
par(bg="white")
plot(df_weekend$temperatura, df_weekend$vanzari, main = "Weekend vs zilele saptamanii", col="#00FF00")

```

```r
par(bg="white")
plot(df_weekend$temperatura, df_weekend$vanzari, main = "Weekend vs zilele saptamanii", col="#00FF00")
plot(df_zile_lucratoare$temperatura, df_zile_lucratoare$vanzari, main = "Weekend vs zilele saptamanii", col="#0000FF")
```

```r
par(bg="white")
plot(df_weekend$temperatura, df_weekend$vanzari, main = "Weekend vs zilele saptamanii", col="#00FF00")
points(df_zile_lucratoare$temperatura, df_zile_lucratoare$vanzari, main = "Weekend vs zilele saptamanii", col="#0000FF")
```

###  V2 - folosind ggplot2. 

```r
head(df)
```

```r
weekend <- c('sambata','duminica')
weekend
```

```r
df <- df %>% mutate(perioada= ifelse(zi_saptamana %in% weekend,  "weekend", "zi normala"))
head(df)
```

```r
ggplot(data = df, aes(x = temperatura, y = vanzari)) + geom_point() + labs(x = "temperatura", y = "vanzari", title= "Vanzari in functie de temperatura" ) + theme_bw()
```

##### Functia scale_color_manual

```r
ggplot(data = df, aes(x = temperatura, y = vanzari, color=perioada )) + geom_point() + labs(x = "temperatura", y = "vanzari" , title= "Vanzari in functie de temperatura") + scale_colour_manual(values = c("weekend" = "#00FF00", "zi normala" = "#0000ff")) + theme_bw()
```
