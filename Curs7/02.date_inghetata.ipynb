```python
LUNI_DE_VARA = ['iunie',
       'iulie', 'august']
```

```python
colors = ["#FF0000" if luna in LUNI_DE_VARA else "#0000FF" for luna in monthly_total_sales.index]
colors
```

```python
plt.rcParams["xtick.color"]
```

```python
plt.rcParams["xtick.color"] = "red"
```

```python
plt.rcParams["figure.figsize"] # [6.4, 4.8]
```

```python

plt.xticks(rotation=45);
plt.figure(figsize=(6.4 * 2, 4.8));
plt.bar(monthly_total_sales.index, monthly_total_sales);
```

```python
plt.rcParams["xtick.color"] = "black"
```

# Exercitiul 6
## Împarte lunile în 4 anotimpuri și creează un pie chart cu procentul total de vânzări pe fiecare anotimp.


```python
df
```

```python
df["luna"].unique()
```

```python
ANOTIMPURI = ["iarna", "primavara", "vara", "toamna"]
LUNI_ANOTIMP = ['decembrie', 'ianuarie', 'februarie', 'martie', 'aprilie', 'mai', 'iunie',
       'iulie', 'august', 'septembrie', 'octombrie', 'noiembrie']
```

```python
LUNI_MAPATE = { luna:ANOTIMPURI[index // 3] for index, luna in enumerate(LUNI_ANOTIMP) }
LUNI_MAPATE
```

```python
df["anotimp"] = df["luna"].map(LUNI_MAPATE)
df
```

```python
sezon_total_sales = df.groupby("anotimp")["vanzari"].sum()
sezon_total_sales
```

```python
plt.pie(sezon_total_sales, labels=sezon_total_sales.index, autopct="%1.2f%%");
```
