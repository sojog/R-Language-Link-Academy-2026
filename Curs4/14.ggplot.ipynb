```r
library("ggplot2")

x <- 101:110
y <- 91:100

df <- data.frame(x, y)

"Exista 2 functii in packet - qplot, ggplot"

ggplot(df, aes(x, y))
```

```r
ggplot(df, aes(x, y)) + geom_point()
```

```r
ggplot(df, aes(x, y)) + geom_boxplot()
```

```r
ggplot(df, aes(x, y)) + geom_line()
```

```r
ggplot(df, aes(x, y)) + geom_bar()
```
